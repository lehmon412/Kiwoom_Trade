```python
def get_stock_list(market_list):
    now = datetime.now()
    today_date = now.strftime('%Y%m%d')
    stock_list_df = pd.DataFrame()
    for market_nm in market_list:
        stock_list = stock.get_market_ticker_list(today_date, market=market_nm)
        
        for stocks in stock_list:
            corp_name = stock.get_market_ticker_name(stocks)
            temp_df = pd.DataFrame({'ticker':stocks,
                            'corp_name':corp_name,
                            'marke': market_nm
                            }, index = [0])
            stock_list_df = pd.concat([stock_list_df,temp_df])
    stock_list_df = stock_list_df.reset_index(drop = True)
    return stock_list_df
```

```python
import exchange_calendars as ecals

def Convert_To_Workday(date):
   shDate = date.strftime('%Y-%m-%d')
   XKRX = ecals.get_calendar("XKRX") # 한국 코드
   if not XKRX.is_session(shDate):
      shiftedDate = XKRX.previous_close(pd.Timestamp(date))
      shiftedWorkday = shiftedDate.strftime('%Y-%m-%d')
   else:
      shiftedWorkday = shDate
   
   return shiftedWorkday
```

```python

def preprocess_stock_dataframe(ticker, start_dt, end_dt):
    # 국장만 적용되도록 설정
    MA_Start = datetime.strptime(start_dt, "%Y-%m-%d") - timedelta(days=200)
    MA_End = datetime.strptime(end_dt, "%Y-%m-%d") + timedelta(days=1)
    df = pdr.get_data_yahoo(ticker, MA_Start, MA_End)
    
    # 60, 120 이평선 계산
    df['MA_60'] = df['Close'].rolling(60).mean().round()
    df['MA_120'] = df['Close'].rolling(120).mean().round()
    df['Close_max_60'] = df['Close'].rolling(60).max()
    
    #등락률 계산
    df['rate'] = round((df['Close'] - df['Close'].shift(1)) / df['Close'].shift(1) * 100, 2)
    #null값 쿼리로 null 문자로 처리
    df.fillna('NULL', inplace=True)
    
    return df
```

```python
def generate_insert_queries(df, ticker,start_dt):
    
    queries = []
    start_date = pd.to_datetime(start_dt)

    for index, row in df.iterrows():
        if index >= start_date:
            MARKET_date = index.strftime('%Y-%m-%d')
            values = ', '.join([f"'{ticker}'", f"TO_DATE('{MARKET_date}', 'YYYY-MM-DD')", str(row['Open']), 
                                str(row['Close']), str(row['High']), str(row['Low']), str(row['rate']), 
                                str(row['Volume']), str(row['MA_60']), str(row['MA_120'])])
            query = f"INSERT INTO DAILYMARKET (STOCK_KEY, MARKET_DT, OPEN_PRICE, CLOSE_PRICE, HIGH_PRICE, LOW_PRICE, DAILY_RATE, TRADING_VOLUME, MA_60, MA_120) VALUES ({values})"
            
            queries.append(query)

    return queries
```
